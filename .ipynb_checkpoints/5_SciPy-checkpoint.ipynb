```python
import numpy as np
import matplotlib.pylab as plt
import scipy
from scipy.optimize import curve_fit
from scipy.optimize import minimize
%matplotlib inline
```

```python
def fun(x, a,b,c,d):
    return a*x**3 + b * x**2 + c * x + d
```

```python
#Возьмем дисперсию побльше, чтобы разница была видна
scale = 35
x = np.linspace(-5, 5, 500)
y =  np.apply_along_axis(fun, 0, x, *[1,2,-3,2])
y_noise = y + np.random.normal(scale=scale, size=500)
plt.figure(figsize=(10,8))
plt.grid()
plt.plot(x,y)
plt.plot(x, y_noise, 'x')
```

```python
#Считаем коэфициэнты по данным с шумум
popt, pcov = curve_fit(fun, x, y_noise)
#Считаем экстремумы для оригинальных данных и данных с шумом
min_noise = minimize(lambda x: fun(x, *popt), 0).x[0]
min_old = minimize(lambda x: fun(x, *[1,2,-3,2]), 0).x[0]
max_noise = minimize(lambda x: -fun(x, *popt), 0).x[0]
max_old = minimize(lambda x: -fun(x, *[1,2,-3,2]), 0).x[0]
#Все строим
plt.figure(figsize=(16,12))
plt.grid()
plt.plot(x, y_noise, 'x', label='noise')
plt.plot(x,y, label='original')
plt.plot(x, fun(x, *popt), label='predict')
plt.plot([min_old, max_old], [fun(min_old, *[1,2,-3,2]), fun(max_old, *[1,2,-3,2])], 'o', 
         label='extremum oroginal')
plt.plot([min_noise, max_noise], [fun(min_noise, *popt), fun(max_noise, *popt)], 'o',
         label='extremum predict')
plt.legend()
```
