```python
class PolynomialRegression(object):
    
    def __init__(self, max_degree=1, t=0):
        self.max_degree = max_degree
        self.w = np.zeros((max_degree+1))
        self.t = t
        
    def fit(self, points_x, points_y):
        X = np.array(list((map(lambda x: np.array(list(map(lambda i: x**i, range(self.max_degree+1))))
                               ,points_x))))
        #print(X)
        print(np.identity(len(np.transpose(X) @ X)) * self.t)
        self.w = np.linalg.inv(np.transpose(X) @ X + np.identity(len(np.transpose(X) @ X)) * self.t) @ np.transpose(X) @ points_y
        #print(self.w)
        #print(self.penalty)
        return self
    
    def predict(self, points_x):
        values = np.zeros((len(points_x)))
        for i in range(len(points_x)):
            values[i] = sum(np.array(list(map(lambda t: t[0] * points_x[i]**t[1], 
                                              zip(self.w,range(self.max_degree+1)))))) 
        return values
```

```python
np.random.seed(42)
points_x = np.random.uniform(-10, 10, size=7)
# we use list comprehesion but think about how to write it using np.array operations
points_y = np.array([4 - x + x ** 2 + 0.1 * x ** 3 + np.random.uniform(-20, 20) for x in points_x])
```

```python
def plot_model(max_degree):
    plt.figure(figsize=(10, 5))
    plt.scatter(points_x, points_y)
    model = PolynomialRegression(max_degree).fit(points_x, points_y)
    all_x = np.arange(-10, 10.1, 0.1)
    plt.plot(all_x, model.predict(all_x))
    plt.grid()
    plt.show()
```

```python
for i in range(10):
    plot_model(i)
```

```python
def plot_model_2(max_degree, t=0):
    plt.figure(figsize=(10, 5))
    plt.scatter(points_x, points_y)
    model = PolynomialRegression(max_degree=max_degree, t=t).fit(points_x, points_y)
    all_x = np.arange(-10, 10.1, 0.1)
    plt.plot(all_x, model.predict(all_x))
    plt.grid()
    plt.show()
```

```python
for i in range(10):
    plot_model_2(i, 2*10**(6))
```
