```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import matplotlib as mp
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
%matplotlib inline
```

## 2

```python
x = np.linspace(0.01, 4, 100)

f_arr = [lambda x: 10 * np.sin(x), lambda x: x * np.log(x), lambda x: 2 * x**2]
f_name = ["$10\\sin{x}$", "$x\\ln{x}$","$2x^2$"]
plt.figure(figsize = (10,10))
plt.xlabel("$x$")
plt.ylabel("$y$")
for f,name in zip(f_arr,f_name):
    plt.plot(x, f(x), label=name)
plt.legend()
plt.savefig('graph.png', format='png', dpi=100)
```

## 3

```python
plt.rcParams['figure.figsize'] = 12, 10
x = np.linspace(0.01, 4, 100)
f_arr = [lambda x: x**2, lambda x: x**3, lambda x: np.exp(x), lambda x: np.log(x)]
f_name = ["$x^2$", "$x^3$", "$e^{x}$", "$\\ln{x}$"]
p = [221, 222, 223, 224]
for i in range(len(p)):
    plt.subplot(p[i])
    plt.plot(x, f_arr[i](x))
    plt.xlabel("$x$")
    plt.ylabel("$y$")
    plt.title(f_name[i], fontsize = 10)
```

```python
def fun(x, y):
    return 3 * x * y + x - 2 * y
N = 100
x = np.linspace(0, 5, N)
y = np.linspace(0, 5, N)
data = np.zeros(shape=(N, N))
for i in range(N):
    for j in range(N):
        data[i][j] = fun(x[i], y[j])
# На семинаре был seaborn
sns.heatmap(data, xticklabels=False, yticklabels=False)
```

```python
def fun(x, y):
    return 3 * x * y + x - 2 * y
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)
xv, yv = np.meshgrid(x, y)
zv = 3 * xv * yv + xv - 2 * yv
fig = plt.figure()
axes = Axes3D(fig)

axes.plot_surface(xv, yv, zv)
```
